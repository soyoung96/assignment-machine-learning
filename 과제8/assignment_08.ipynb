# Logistic Regression 

## import library

## load training data

```python
fname_data = 'assignment_08_data.csv'

data = np.genfromtxt(fname_data, delimiter=',')
num_data = data.shape[0]

point_x = np.zeros(num_data)
point_y = np.zeros(num_data)
label   = np.zeros(num_data)

for i in range(num_data):

    point_x[i]  = data[i,0]
    point_y[i]  = data[i,1]
    label[i]    = data[i,2]
```

## define linear regression function with inputs $\theta = (\theta_0, \theta_1, \theta_2)$ and ${\rm point} = (1, x, y)$

```python
def linear_regression(x,y,theta):
    
    num_data = x.shape[0]
    
    one=np.ones(shape=(num_data,))
    value = theta[0]*one + theta[1]*x + theta[2]*y
    
        

    return value
```

## define sigmoid function with input $x$

```python
def sigmoid(x):

    z=1/(1+np.exp(-x))

    return z
```

## define loss function for the logistic regression

```python
def compute_loss(x,y,theta,l):
    
    num_data = x.shape[0]
    k = linear_regression(x,y,theta)
    h = sigmoid(k)
    loss = (((-l*np.log(h))-((1-l)*np.log(1-h))).sum())/num_data
    
    
    return loss
```

## define gradient vector for the model parameters $\theta = (\theta_0, \theta_1, \theta_2)$

```python
def compute_gradient(x,y,theta,l):
    
    num_data = x.shape[0]
    one=np.ones(shape=(num_data,))
    X = np.c_[one,x,y]
    k = linear_regression(x,y,theta)
    h = sigmoid(k)
    for i in range(num_data):
        X[i]=(h[i]-l[i])*X[i]
    gradient = X.sum(axis = 0)/num_data

    return gradient
```

## gradient descent for the model parameters $\theta = (\theta_0, \theta_1, \theta_2)$

```python
num_iteration   = 5000  # USE THIS VALUE for the number of gradient descent iterations 
learning_rate   = 0.001 # USE THIS VALUE for the learning rate
theta           = np.array((0, 0, 0))   # USE THIS VALUE for the initial condition of the model parameters
theta_iteration = np.zeros((num_iteration, theta.size))
loss_iteration  = np.zeros(num_iteration)
```

## iterations for the gradient descent

```python
for i in range(num_iteration):
    theta  = theta - learning_rate * compute_gradient(point_x, point_y, theta, label)
   
    loss    = compute_loss(point_x, point_y, theta, label)

    theta_iteration[i] = theta
    loss_iteration[i]   = loss
    
    print("iteration = %4d, loss = %5.5f" % (i, loss))
```

## plot the results

```python
def plot_loss_curve(loss_iteration):

    plt.figure(figsize=(8,6))   
    plt.title('loss')


    plt.plot(list(range(num_iteration)), list(loss_iteration), color = 'red')


    plt.xlabel('iteration')
    plt.ylabel('loss')

    plt.tight_layout()
    plt.show()
```

```python
def plot_data(data):
    
    plt.figure(figsize=(8,8))
    plt.title('training data')
    
    point_0_x=[]
    point_1_x=[]
    point_0_y=[]
    point_1_y=[]
    
    for i in range(num_data):
        if (data[i,2]==0):
            point_0_x.append(data[i,0])
            point_0_y.append(data[i,1])
        elif (data[i,2]==1):
            point_1_x.append(data[i,0])
            point_1_y.append(data[i,1])
            
            
    
    
    plt.scatter(point_0_x, point_0_y, marker='o', color='blue',s=7, label='class = 0')
    plt.scatter(point_1_x, point_1_y, marker='o', color='red',s=7,label='class = 1')


    plt.axis('equal')
    plt.legend(loc='upper right')
    plt.tight_layout()
    plt.show()
```

```python

def plot_model_parameter(theta_iteration):

    plt.figure(figsize=(8,6))
    plt.title('model parameter')


    plt.plot(list(range(num_iteration)), list(theta_iteration[:,0]), color = 'red',label='θ0')
    plt.plot(list(range(num_iteration)), list(theta_iteration[:,1]), color = 'green',label='θ1')
    plt.plot(list(range(num_iteration)), list(theta_iteration[:,2]), color = 'blue',label='θ2')

    
    plt.xlabel('iteration')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

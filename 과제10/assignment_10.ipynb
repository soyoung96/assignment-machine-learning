## define feature function

```python
def feature_function(x, y):
    
    #3차 곡선 즉 1,X^3,Y
    one=np.ones(shape=(x.shape[0],))
    feature = np.c_[one,x,x**2,x**3,y]
    return feature
```

## define regression function based on the feature function

```python
def regression_function(theta, feature):
    
    value = np.dot(feature,theta)
    return value
```

## define regularization function on the model parameters

```python
def regularization_function(theta):
    
    value = (theta ** 2).sum(axis=0)
    
    
    return value
```

## define sigmoid function

```python
def logistic_function(x):
    
    z=1/(1+np.exp(-x))
    
    return z
```

## define loss function where $\alpha$ is a weight for the quadratic regularization term (Note that you need to add a small number (np.finfo(float).eps) inside logarithm function in order to avoid $\log( 0 )$)

```python
def compute_loss_feature(theta, feature, label, alpha):
    
    k = regression_function(theta, feature)
    h = logistic_function(k)
    l = label
    regular = regularization_function(theta)
    loss = ((((-l*np.log(h))-((1-l)*np.log(1-h))).sum()) + (alpha*regular/2))/feature.shape[0]
    
    return loss
```

## define gradient vector for the model parameters with the quadratic regularization term whose weight is $\alpha$

```python
def compute_gradient_feature(theta, feature, label, alpha):
    
    num_data = feature.shape[0]
    X = feature
    k = regression_function(theta, feature)
    h = logistic_function(k)
    l=label
    
    gradient=(np.dot((h-l),X)/num_data) + (alpha * (theta))
    
    
    return gradient
```

## compute the accuracy

```python
def compute_accuracy(theta, feature, label):
    
    num_data = feature.shape[0]
    k = regression_function(theta, feature)
    h = logistic_function(k)
    
    ans_num = 0
    
    for i in range(num_data):
        
        if(h[i] >= 0.5):
            ans = 1
        else:
            ans = 0
        if (label[i] == ans):
            ans_num += 1
    
    accuracy = ans_num / num_data
    
    return accuracy
```

## gradient descent for the model parameters $\theta$

```python
num_iteration   = 20000    
learning_rate   = 0.7
alpha           =0.0001

theta1           = np.array((0, 0, 0,0,0)) 


feature1 = feature_function(data_train[:,0], data_train[:,1])
feature2 = feature_function(data_test[:,0], data_test[:,1])
```

```python
theta_iteration             = np.zeros((num_iteration, theta1.size))
loss_iteration_train        = np.zeros(num_iteration)
loss_iteration_test         = np.zeros(num_iteration)
accuracy_iteration_train    = np.zeros(num_iteration)
accuracy_iteration_test     = np.zeros(num_iteration)
```

```python
for i in range(num_iteration):
 
    theta1  = theta1 - learning_rate * compute_gradient_feature(theta1, feature1, data_train[:,2],alpha)
   
    loss1    = compute_loss_feature(theta1, feature1, data_train[:,2],alpha)
    loss2    = compute_loss_feature(theta1, feature2, data_test[:,2],alpha)
    
    accuracy1    = compute_accuracy(theta1, feature1, data_train[:,2])
    accuracy2    = compute_accuracy(theta1, feature2, data_test[:,2])
    
   
    theta_iteration[i] = theta1
    
    loss_iteration_train[i]   = loss1
    loss_iteration_test[i]   = loss2
    
    accuracy_iteration_train[i]    = accuracy1
    accuracy_iteration_test[i]     = accuracy2
    
    print("[%d] loss(train) = %5.5f, loss(test) = %5.5f, accuracy(train) = %5.5f , accuracy(test) = %5.5f" % (i,loss1,loss2,accuracy1,accuracy2))
    

theta_optimal = theta1
#3887
```

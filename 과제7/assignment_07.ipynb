## import library

```python
import numpy as np
import matplotlib.image as img
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from mpl_toolkits.mplot3d import Axes3D
```

## load point data for training and testing

```python
fname_data  = 'assignment_07_data.csv'

data        = np.genfromtxt(fname_data, delimiter=',')
num_data    = data.shape[0]

xx          = np.zeros(num_data)
yy          = np.zeros(num_data)
zz          = np.zeros(num_data)

for i in range(num_data):

    xx[i] = data[i,0]
    yy[i] = data[i,1]
    zz[i] = data[i,2]
    
```

## plot the data in the three dimensional space

```python
fig = plt.figure(figsize=(12, 8))

ax1 = plt.subplot(111, projection='3d')
plt.title('data')

ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_zlabel('$z$')

ax1.scatter(xx, yy, zz, marker='o', color='blue', alpha=0.2)

plt.tight_layout()
plt.show()
```

## compute the loss function

```python
def compute_residual(theta, x, y, z):
    # ===== FILL UP THE CODE =====
    
    num_data = x.shape[0]
    
    one=np.ones(shape=(num_data,))
    X = np.c_[one,x,y]
    
    residual = (np.dot(X,theta)) - z
    
    
    
    
    # ============================
    return residual, num_data
```

```python
def compute_loss(theta, x, y, z):
    # ===== FILL UP THE CODE =====
    
    residual, num_data = compute_residual(theta,x,y,z)
    residual_2 = residual * residual
    sum_residual_2=residual_2.sum()
    loss= sum_residual_2 / (2*num_data)
    
    # ============================
    return loss
```

## compute the gradient for each model parameter (DO NOT COMPUTE THE GRADIENT FOR EACH MODEL PARAMETER, BUT DO COMPUTE THE GRADIENT OF THE MODEL PARAMTER VECTOR)

```python
def compute_gradient(theta, x, y, z):
    # ===== FILL UP THE CODE =====
    one=np.ones(shape=(num_data,))
    X = np.c_[one,x,y]
    Xw_z = np.dot(X,theta) - z
    
    grad = np.dot(X.T,Xw_z) * 1/num_data
    
    # ============================
    return grad
```

## gradient descent for each model parameter

```python
num_iteration   = 1000
learning_rate   = 0.01

theta           = np.array((0, 0, 0))
theta_iteration = np.zeros((num_iteration, theta.size))
loss_iteration  = np.zeros(num_iteration)

for i in range(num_iteration):
    # ===== FILL UP THE CODE =====

    theta  = theta - learning_rate * compute_gradient(theta, xx, yy, zz)
   
    loss    = compute_loss(theta, xx, yy, zz)

    theta_iteration[i] = theta
    loss_iteration[i]   = loss

    # ============================
    print("iteration = %4d, loss = %5.5f" % (i, loss))
```

```python
f = theta[0] + theta[1] * xx + theta[2] * yy
```

## plot the results

```python
def plot_loss_curve(loss_iteration):

    plt.figure(figsize=(8,6))
    plt.title('loss')
    # ===== FILL UP THE CODE =====
    plt.plot(list(range(num_iteration)), list(loss_iteration), color = 'red')



    # ============================
    plt.tight_layout()
    plt.show()
```

```python
def plot_data(xx, yy, zz):
    
    fig = plt.figure(figsize=(12, 8))
    ax  = plt.subplot(111, projection='3d')
    plt.title('data')
    # ===== FILL UP THE CODE =====


    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$z$')

    ax.scatter(xx, yy, zz, marker='o', color='blue')


    

    # ============================    
    plt.tight_layout()
    plt.show()
```

```python

def plot_model_parameter(theta_iteration):

    plt.figure(figsize=(8,6))
    plt.title('model parameter')
    # ===== FILL UP THE CODE =====


    plt.plot(list(range(num_iteration)), list(theta_iteration[:,0]), color = 'red')
    plt.plot(list(range(num_iteration)), list(theta_iteration[:,1]), color = 'green')
    plt.plot(list(range(num_iteration)), list(theta_iteration[:,2]), color = 'blue')


    # ============================
    plt.tight_layout()
    plt.show()
```

```python
plot_loss_curve(loss_iteration)
```

```python
plot_model_parameter(theta_iteration)
```

```python
# ===== FILL UP THE CODE =====

grid_X  = np.arange(-10, 10, 0.1)
grid_Y  = np.arange(-10, 10, 0.1)

XX, YY = np.meshgrid(grid_X,grid_Y)

ZZ=list()
for i in range(len(XX)):
    for j in range(len(XX)):
        ZZ.append(theta[0] + theta[1] * XX[i,j] + theta[2] * YY[i,j])
ZZ = np.asarray(ZZ).reshape(XX.shape)
    

# ============================

def plot_surface(XX, YY, ZZ, xx, yy, zz):
    fig = plt.figure(figsize=(8,8))
    ax  = fig.add_subplot(111, projection='3d')
    plt.title('regression surface')

    ax  = plt.axes(projection='3d')
    # ===== FILL UP THE CODE =====

    
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$z$')

    
    ax.plot_surface(np.asarray(XX), np.asarray(YY), ZZ)
    
    ax.scatter(xx, yy, zz, marker='o', color='blue')

    # ============================
    plt.tight_layout()
    plt.show()
```

```python
plot_surface(XX, YY, ZZ, xx, yy, zz)
```
